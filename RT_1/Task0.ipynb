### Спектр одиночного прямоугольного импльса

$x(t) = 1 \hspace{5mm} |t| < \tau$

$X(f) = \tau \frac{\sin(\pi f \tau)}{\pi f \tau}$

Грфик:

```python
from Plot import MyPlot
from Plot import PlotFunction 

import numpy as np

tau = 1

def spectrum(f):
    return tau * np.sin(np.pi * f * tau) / (np.pi * f * tau) 

def main():
    func = PlotFunction()
    func.create_continuous_function(spectrum, -5 / tau, 5 / tau)

    plot = MyPlot()
    plot.add_function(func)
    plot.config_plot('Спектр прямоугольного импульса', 'Частота', 'Амплитуда')
    plot.draw_all()
    MyPlot.show_all()

if __name__ == '__main__':
    main()
```

###  Спектр синусоидального сигнала

Синусоида раскладывается на две мнимые экспоненты, таким образом ее спектр будет состоять из двух симметрично расположенных дельта функций.

$x(t) = \sin(\omega t)$

$X(f) = -\frac{i}{2} \delta(f - \frac{\omega}{2 \pi}) + \frac{i}{2} \delta(f + \frac{\omega}{2 \pi})$

График:

```python
from Plot import MyPlot
from Plot import PlotFunction 

import numpy as np

omega = 1 

def main():
    left_func = PlotFunction()
    left_func.set_arrayX([-omega / (2 * np.pi), -omega / (2 * np.pi)])
    left_func.set_arrayY([1, 0])
    
    right_func = PlotFunction()
    right_func.set_arrayX([omega / (2 * np.pi), omega / (2 * np.pi)])
    right_func.set_arrayY([-1, 0])

    plot = MyPlot()
    plot.add_function(left_func)
    plot.add_function(right_func)
    plot.config_plot('Спектр синусоиы', 'Частота', 'Комплексная амплитуда')
    plot.draw_all()
    MyPlot.show_all()

if __name__ == '__main__':
    main()
```

### Спектр треуольного импульса 

Спектр симметричного прямоугольного импульса:

$K(f) = \frac{\tau}{2} \frac{\sin^2(\pi f \tau / 2)}{(\pi f \tau / 2)^2}$

```python
from Plot import MyPlot
from Plot import PlotFunction 

import numpy as np

tau = 1

def spectrum(f):
    return tau / 2 * (np.sin(np.pi * f * tau / 2)**2) / ((np.pi * f * tau / 2)**2) 

def main():
    func = PlotFunction()
    func.create_continuous_function(spectrum, -5 / tau, 5 / tau)

    plot = MyPlot()
    plot.add_function(func)
    plot.config_plot('Спектр треугольного импульса', 'Частота', 'Амплитуда')
    plot.draw_all()
    MyPlot.show_all()

if __name__ == '__main__':
    main()
```
